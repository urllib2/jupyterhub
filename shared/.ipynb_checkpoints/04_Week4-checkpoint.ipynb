## ğŸ” Understanding Robot State Publisher

**robot_state_publisher** requires:
1. **URDF** as `robot_description` parameter
2. **Joint states** from `/joint_states` topic

### Check Parameters:
```bash
ros2 param list /robot_state_publisher
ros2 param get /robot_state_publisher robot_description
```

### Important Distinction:
- **Joint State**: Current position/velocity of ONE joint
- **TF**: Relationship between TWO frames (transformation)

## ğŸ“¦ Creating a ROS 2 Package

### Why Package?
- Organize URDF, launch files, configurations
- Share code with others
- Install in ROS 2 workspace

### Create New Workspace:
```bash
# Create workspace
mkdir -p ~/my_robot_ws/src

# Update .bashrc (comment old workspace)
# source ~/ros2_ws/install/setup.bash  # â† Comment this
source ~/my_robot_ws/install/setup.bash  # â† Add this
```

### Create Description Package:
```bash
cd ~/my_robot_ws/src
ros2 pkg create my_robot_description --build-type ament_cmake
cd my_robot_description
rm -r include src  # No C++ code needed
```

### Install Folders in CMakeLists.txt:
```cmake
install(
  DIRECTORY urdf rviz launch
  DESTINATION share/${PROJECT_NAME}/
)
```

### Update Xacro Includes:
```xml
<xacro:include filename="$(find my_robot_description)/urdf/common_properties.xacro" />
```

## ğŸš€ XML Launch File

**File:** `display.launch.xml`

```xml
<launch>
  <!-- Define paths -->
  <let name="urdf_path" 
       value="$(find-pkg-share my_robot_description)/urdf/my_robot.urdf.xacro" />
  <let name="rviz_config" 
       value="$(find-pkg-share my_robot_description)/rviz/urdf_config.rviz" />

  <!-- robot_state_publisher -->
  <node pkg="robot_state_publisher" exec="robot_state_publisher">
    <param name="robot_description" 
           value="$(command 'xacro $(var urdf_path)')" />
  </node>

  <!-- joint_state_publisher_gui -->
  <node pkg="joint_state_publisher_gui" 
        exec="joint_state_publisher_gui" />

  <!-- RViz -->
  <node pkg="rviz2" exec="rviz2" args="-d $(var rviz_config)" />
</launch>
```

### Launch It:
```bash
# Build
cd ~/my_robot_ws
colcon build --packages-select my_robot_description

# Source
source install/setup.bash

# Launch
ros2 launch my_robot_description display.launch.xml
```

---

## âš–ï¸ XML vs Python Launch Files

**XML (Recommended):**
- 15 lines for our example
- Simple, readable
- Easy to maintain

**Python:**
- 38+ lines for same functionality
- More complex syntax
- Better for complex logic

**Our choice:** XML for simplicity
