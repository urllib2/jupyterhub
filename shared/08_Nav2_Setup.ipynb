## 🏗️ ROS2 Workspace Architecture

Before modifying Turtlebot3, you need to understand how ROS2 workspaces work. You never modify the global installation — you create an overlay workspace that overrides specific packages.

![ROS2 workspace overlay and underlay](images/nav2_slide09.jpg)

- **Underlay** — `/opt/ros/jazzy` — global read-only installation
- **Overlay** — `~/turtlebot3_ws/src` — your custom development workspace

When you `colcon build` and `source install/setup.bash`, your overlay takes precedence.
