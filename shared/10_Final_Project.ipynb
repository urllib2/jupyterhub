## Step 1 — Build Your World

Create a custom Gazebo environment using the world editor. Requirements:
- At least 3 rooms connected by corridors
- Corridors must be at least **1.2m wide** (Turtlebot3 waffle is 0.28m — leave room for the inflation radius)
- At least one U-shaped path that requires waypoints

```bash
gazebo --verbose -s libgazebo_ros_init.so -s libgazebo_ros_factory.so
```

Save as `my_world.world` and update the launch file.

---

## Step 2 — Create the Overlay Workspace

```bash
mkdir -p ~/turtlebot3_ws/src
cd ~/turtlebot3_ws/src
git clone -b jazzy https://github.com/ROBOTIS-GIT/turtlebot3_simulations.git
cd ~/turtlebot3_ws
colcon build --symlink-install
source install/setup.bash
```

Modify `turtlebot3_world.launch.py` to use your world and set the spawn position.

---

## Step 3 — Generate the Map with SLAM

Launch the robot in your world:
```bash
ros2 launch turtlebot3_gazebo turtlebot3_world.launch.py
```

Start SLAM:
```bash
ros2 launch turtlebot3_cartographer cartographer.launch.py use_sim_time:=True
```

Drive through every room:
```bash
ros2 run turtlebot3_teleop teleop_keyboard
```

Save the map:
```bash
ros2 run nav2_map_server map_saver_cli -f ~/my_map
```

---

## Step 4 — Post-Process the Map

Open `~/my_map.pgm` in GIMP:
- Seal all wall gaps with the 1-pixel black pencil
- Clear grey noise in free space with white
- Export and overwrite

Verify the cleaned map looks correct before moving to Nav2.

---

## Step 5 — Launch Nav2 and Navigate

Launch Nav2 with your cleaned map:
```bash
ros2 launch turtlebot3_navigation2 navigation2.launch.py \
  use_sim_time:=True \
  map:=$HOME/my_map.yaml
```

In RViz:
1. Set the **2D Pose Estimate** — tell Nav2 where the robot starts
2. Set a **Nav2 Goal** — click anywhere on the map
3. Watch the robot plan and execute the path

For complex routes, use **Waypoint Mode** to break the navigation into segments.

Tune `inflation_radius` in `nav2_params.yaml` if the robot gets blocked in corridors.

---

## 🧠 Final Challenge

Your robot must complete this mission autonomously:

1. Start at position A (defined in your launch file)
2. Navigate to position B (far side of the world, through at least 2 corridors)
3. Navigate back to position A
4. Complete the round trip without hitting any wall

Record a screen capture of the successful navigation in RViz.

**Submission checklist:**
- [ ] Custom Gazebo world with at least 3 rooms
- [ ] Clean map with no grey zones between rooms
- [ ] Nav2 launches without errors
- [ ] Robot completes the A→B→A round trip
- [ ] No wall collisions
