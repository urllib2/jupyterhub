## 🎯 The Core Objective

Nav2 solves one problem: move a robot from Point A to Point B safely, while avoiding static walls, dynamic objects, and humans.



---

## 🗺️ The Two-Step Process

Autonomous navigation always follows two phases:

1. **Map the world** — drive manually, use SLAM to generate a static floorplan
2. **Navigate** — run Nav2 to autonomously route through the known map while avoiding dynamic obstacles



You already completed Phase 1 in Lesson 6. This lesson covers Phase 2.

---

## 🧠 Nav2 Architecture — Simplified

Nav2 is an input/output engine: it consumes spatial data, computes a path, and outputs motor commands.

![Simplified Nav2 architecture](images/nav2_slide04.jpg)

**Inputs:**
- **TF tree** — where every part of the robot is
- **Map** — the saved SLAM map
- **Sensor data** — LiDAR scans for dynamic obstacle detection

**The brain:**
- **Planner Server** — computes the global path
- **Controller Server** — executes the path with velocity commands

**Output:**
- Velocity commands on `/cmd_vel`

---

## 📐 The TF Challenge

Nav2 needs to know exactly where every part of the robot is at all times. A robot is a collection of moving parts — transforms (TF) continuously compute precise 3D position and rotation for every component.



---

## 🌳 The Required TF Tree

For Nav2 to work, the TF tree must be complete and unbroken — from the global map all the way down to the LiDAR sensor.



If any link in this chain is missing, Nav2 cannot localize the robot and navigation fails.

---

## 🧭 Map Frame vs Odom Frame

Nav2 uses two position sources simultaneously — each with different strengths:



- **Map frame** — comes from SLAM/laser scans. Long-term precision but can jump suddenly when the position is corrected.
- **Odom frame** — comes from wheel odometry. Short-term smoothness but drifts over distance.

Nav2 uses both together to get the best of each.

---

## 🔗 The Dual-Parent Hack

ROS rules say a frame can only have ONE parent. Nav2 solves the map/odom problem by linking them linearly:



The chain `map → odom → base_footprint` absorbs the sudden map corrections at the `map→odom` level, keeping `odom→base_footprint` smooth.

---

## 🤖 Behavior Trees — The Orchestrator

Nav2 uses behavior trees (BT) to orchestrate planning, execution, and recovery. Logic trees evaluate states continuously to shift between planning, executing, and recovering.



---

## 🏗️ The Complete Nav2 Architecture

Putting it all together:

![Complete Nav2 architecture](images/nav2_slide20.jpg)

The Behavior Tree Navigator orchestrates everything — it calls the Planner Server for a global path, the Controller Server to follow it, and the Recovery Server when something goes wrong.

## ✅ What you learned

- Nav2 moves a robot from A to B by combining a global path planner and a local controller
- It requires a complete TF tree: `map → odom → base_footprint → base_link → sensors`
- Map frame gives long-term precision, Odom frame gives short-term smoothness
- The dual-parent chain absorbs map correction jumps
- Behavior trees orchestrate planning, execution, and recovery
